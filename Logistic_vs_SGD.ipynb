```python
from sklearn.metrics import mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor

x = np.array([2,4,6,8,10]).reshape(-1,1)
y = np.array([55,60,70,80,88])
x,y

```

```python
model_d = LinearRegression()
model_d.fit(x,y)
```

```python
print(model_d.coef_[0])
print(model_d.intercept_)
```

```python
prediction = model_d.predict(x)
prediction
```

```python
lr_mse = mean_squared_error(y,prediction)
lr_mse
```

```python
plt.scatter(x,y)
plt.plot(x,prediction,color='red')
plt.grid(True)
plt.show()
```

# **SGD**

```python
sgd = SGDRegressor(max_iter=1000,tol=1e-3)
sgd.fit(x,y)

print(sgd.coef_[0])
print(sgd.intercept_)
```

```python
pre = sgd.predict(x)

sgd_mse = mean_squared_error(y, pre)
sgd_mse
```

```python
plt.scatter(x,y)
plt.plot(x,pre,color='red')
plt.grid(True)
plt.show()
```

# ***MAE , MSE , RMAE , R^2***

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# SGD
mae = mean_absolute_error(y,pre)
mse = mean_squared_error(y,pre)
rmse = np.sqrt(mean_squared_error(y,pre))
r2_s = r2_score(y,pre) # logistic
print('MAE: ',mae)
print('MSE: ',mse)
print('RMSE: ',rmse)
print("R2_SCORE: ", r2_s)
```

```python
# logistic
mae = mean_absolute_error(y,prediction)
mse = mean_squared_error(y,prediction)
rmse = np.sqrt(mean_squared_error(y,prediction))
r2_s = r2_score(y,prediction)
print('MAE: ',mae)
print('MSE: ',mse)
print('RMSE: ',rmse)
print("R2_SCORE: ", r2_s)
```
