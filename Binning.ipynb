```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("heart.csv")
df.head()
```

```python
df["Age_bin"] = pd.cut(
    df["Age"],
    bins=[0,40,60,100],
    labels=["Young","Middle Age", "Old"]
)
```
